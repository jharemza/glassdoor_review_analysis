```python
#import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
filelist = ["akuna",
            "citadel", 
            "ctc", 
            "deshaw", 
            "drw",
            "fiverings", 
            "hrt", 
            "imc", 
            "jane_street", 
            "millenium", 
            "oldmissioncapital", 
            "optiver", 
            "quantlab",
            "sig",
            "tmg", 
            "tower",
            "two_sigma"
           ]

df_list =[]

for file in filelist:
    file = pd.read_csv("data/" + file + ".csv")
    df_list.append(file)
    
for index, item in enumerate(df_list):
    item['Company'] = filelist[index]
    
full_df = pd.concat(df_list)
```

```python
full_df.groupby('Company')['rating_overall'].describe()
```

```python
sns.set(rc={'figure.figsize':(18,6)})
sns.boxplot(x='Company', y='rating_overall', data=full_df,palette='Set3');
```

```python
sns.set(rc={'figure.figsize':(18,6)})
sns.violinplot(x='Company', y='rating_overall', data=full_df,palette='Set3');
```

```python
full_df.to_csv('data/glassdoor_data.csv',index=False)
```
